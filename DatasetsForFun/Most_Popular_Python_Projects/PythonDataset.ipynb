```python
import pandas as pd
import zipfile
import numpy as np
#from matplotlib import pyplot as plt
import matplotlib.pyplot as plt
```

```python
zf = zipfile.ZipFile('data/Python.zip')
```

```python
zf.namelist()
```

```python
df_python = pd.read_csv(zf.open('popular_python_projects.csv'))
```

```python
df_python.sample(5)
```

```python
df_python.plot(y = "stars", x = "date", kind ="scatter")
```

```python
binwidth = 10000

plt.hist(df_python["stars"], bins=range(min(df_python["stars"]), max(df_python["stars"]) + binwidth, binwidth) )
plt.show()
```
