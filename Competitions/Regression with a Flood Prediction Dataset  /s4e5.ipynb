```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
df_test = pd.read_csv("playground-series-s4e5/test.csv")
df_train = pd.read_csv("playground-series-s4e5/train.csv")
```

```python
#df_train.dtypes
df_train.head(5)
```

```python
df_train_features = df_train.drop(["id","FloodProbability"], axis=1)
df_train_label = df_train["FloodProbability"]
df_train_features

```

```python
#Determine std dev of columns

df_train_features["Total_Feat"] =(df_train_features.sum(axis=1)/len(df_train_features.columns)).round()

df_train_features.std()
df_train_features
```

```python
x ,x_val , y, y_val = train_test_split(df_train_features,df_train_label, test_size=0.3, train_size=0.70, random_state=42, shuffle=True)
```

```python
def predictAnl(model,x,y,x_val, y_val):
    model.fit(x,y)
    y_pred = model.predict(x_val)
    score = r2_score(y_val, y_pred)
    print(score)
    print(model.score(x,y))
```

```python
lr_model = LinearRegression()

predictAnl(lr_model,x,y,x_val, y_val)
```

```python
df_test_features = df_test.drop("id", axis = 1)
df_test_features["Total_Feat"] =(df_train_features.sum(axis=1)/len(df_train_features.columns)).round()
predictions = lr_model.predict(df_test_features)
```

```python
df_submission= pd.DataFrame({
    'id':df_test["id"],
    "FloodProbability": predictions
})
df_submission.to_csv("submission3.csv",index=False)
```
